```python
#converting class column into Numerical form as it is in categorical form 
from sklearn.preprocessing import LabelEncoder 
Label=LabelEncoder()
df[" class"]=Label.fit_transform(df[" class"])
print("accuracy for Logistic regression")
```

```python
# Through pairplot we can identify the correlation between the columns 
sns.pairplot(data=df)
```

```python
#Heatmap defines the correlation 
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),cmap="plasma",annot=True,vmin=-1,vmax=1)
```

```python
#Splitting input and output data
x=df.iloc[:,:-1]
y=df.iloc[:,-1:]
print(x)
```

```python
print(y)
```

```python
#Dividing data for training and testing 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test.head()
```

```python
# Model building using Logistic regression 
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression() 
model1.fit(x_train,y_train)
```

```python
y_pred=model1.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
z1=accuracy_score(y_test,y_pred)
```

```python
#
confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
model2=DecisionTreeClassifier()
model2.fit(x_train,y_train)
```

```python
y_pred2=model2.predict(x_test)
y_pred2
```

```python
z2=accuracy_score(y_test,y_pred2)
```

```python
#
confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_matrix(y_test,y_pred2),annot=True)
```

```python
#SVM
from sklearn.svm import SVC
model3=SVC(kernel="linear")
```

```python
model3.fit(x_train,y_train)
```

```python
y_pred3=model3.predict(x_test)
y_pred3
```

```python
z3=accuracy_score(y_test,y_pred3)
accuracy_score
```

```python
#
confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_matrix(y_test,y_pred3),annot=True)
```

```python
print("Accuracy for the logistic regression model",z1)
print("Accuracy for the Decision Tree model",z2)
print("Accuracy for the SVM model",z3)
```
